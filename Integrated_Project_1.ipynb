Outliernya harus ditangani terlebih dahulu.

```python
def q75(data):
    return data.quantile(0.75)
def q25(data):
    return data.quantile(0.25)
def iqr(data):
    return q75(data) - q25(data)
def top_outlier(data,x):
    return (q75(data[x])+(1.5*iqr(data[x])))
def bot_outlier(data,x):
    return (q25(data[x])-(1.5*iqr(data[x])))
```

```python
max_outlier = top_outlier(clear_game,'all_sales')
min_outlier = bot_outlier(clear_game,'all_sales')
wo_outlier = clear_game.query('@min_outlier<=all_sales<=@max_outlier')
plt.figure(figsize=(30,30))
sb.boxplot(data=wo_outlier, x='platform', y='all_sales',
           order=wo_outlier.pivot_table(index='platform',
                                        values='all_sales',
                                        aggfunc='mean').reset_index().sort_values('all_sales',ascending=False).platform)
```

## Uji Hipotesis

Hipotesis yang akan diuji adalah
1. Apakah Xbox One dengan PC memiliki rating pengguna yang sama?
2. Apakah Rating pengguna genre Action dan Sport berbeda?

Batas error yang dipilih adalah 5%. Karena 5% sudah cukup baik karena ada beberapa nilai yang hilang dan diisi pada Bab 2.

### Jika rating pengguna yang dimaksud adalah rating ESRB.

Setelah saya mencari cara menguji hipotesis data kategorik, saya menemukan metode chi-square, dan ini yang saya temukan untuk uji hipotesisnya.

#### Xbox One dengan PC

```python
xone = wo_outlier1.query('platform=="xone" | platform=="pc"')
xone = pd.crosstab(xone['platform'],xone['rating'])
xone
```

```python
chi2_stat, pv1, dof, expected = scipy.stats.chi2_contingency(xone)
 
```
