Sekarang kita kategorikan model year menjadi 4, dan odometer menjadi 4. Intinya adalah nilai q1 odometer akan mengisi nilai odometer di mobil yang memiliki model year dibawah q1.\
Detailnya adalah\
**model_year < q1 = odometer q1.\
q1 < model_year < q2 = odometer q2.\
dan seterusnya.**

```python
qyear123 = good_year['model_year'].quantile([0.25,0.5,0.75])
qyear123
```

```python
qodo123 = good_odo['odometer'].quantile([0.2,0.4,0.6,0.8])
qodo123
```

Nilai ini yang nantinya akan dimasukkan kedalam fungsi

```python
def isiodo(odo):
    value = odo['model_year']
    if value < qyear123[0.25]:
        return qodo123[0.8]
    if qyear123[0.25] <= value <= qyear123[0.5]:
        return qodo123[0.6]
    if qyear123[0.5] < value <= qyear123[0.75]:
        return qodo123[0.4]
    if value > qyear123[0.75]:
        return qodo123[0.2]
    
```

```python
good_data = car_data
good_data['odometer'].fillna(value='',inplace=True)
manip_data = good_data.loc[~good_data['model_year'].isna() | ~good_data['odometer'].isna()]
manip_data.info()
```

Manip data adalah data yang memiliki salah satu nilai dari odometer atau model_year

```python
manip_data.loc[manip_data['odometer']=='', 'odometer'] = manip_data.apply(isiodo, axis=1)
```

## Memperbaiki Kualitas Data

```python
# Tambahkan nilai waktu dan tanggal pada saat iklan ditayangkan
def tanggal(hari):
    aa = hari['days_listed']
    bb = hari['date_posted']
    return bb + timedelta(days=aa)
```

```python
data_use['last_posted'] = data_use.apply(tanggal, axis =1)
```

```python
# Tambahkan usia kendaraan saat iklan ditayangkan
def age(tahun):
    aa = tahun['model_year']
    bb = tahun['car_age']
    if aa == bb:
        return 1
    return bb - aa
```

```python
data_use['car_age'] = pd.DatetimeIndex(data_use['date_posted']).year
data_use['car_age'] = data_use.apply(age,axis=1)
```

## Mempelajari dan Menangani Outlier

Outlier yang dimaksud adalah pada kolom price,odometer, dan car_age.

```python
def q75(data):
    return data.quantile(0.75)
def q25(data):
    return data.quantile(0.25)
def iqr(data):
    return q75(data) - q25(data)
```

```python
#BATAS ATAS OUTLIER
max_price = (q75(bfr_final_clearance['price'])+(1.5*iqr(bfr_final_clearance['price'])))
max_car_age = (q75(bfr_final_clearance['car_age'])+(1.5*iqr(bfr_final_clearance['car_age'])))
max_odo = (q75(bfr_final_clearance['odometer'])+(1.5*iqr(bfr_final_clearance['odometer'])))
```

```python
#BATAS BAWAH OUTLIER
min_price = (q25(bfr_final_clearance['price'])-(1.5*iqr(bfr_final_clearance['price'])))
min_car_age = (q25(bfr_final_clearance['car_age'])-(1.5*iqr(bfr_final_clearance['car_age'])))
min_odo = (q25(bfr_final_clearance['odometer'])-(1.5*iqr(bfr_final_clearance['odometer'])))
```

```python
# Simpan data tanpa outlier dalam DataFrame yang terpisah
#final_data = bfr_final_clearance.query('price>=min_price'and'price<=@max_price' and )
price_data = bfr_final_clearance.query('price>=@min_price'and'price<=@max_price')
age_data = price_data.query('@min_car_age<=car_age<=@max_car_age')
final_data = age_data.query('@min_odo<=odometer<=@max_odo')
```
